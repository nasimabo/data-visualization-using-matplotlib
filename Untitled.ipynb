```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(11)
y = x **2
print(x,y)
```

```python
plt.scatter(x,y)
```

```python
plt.plot(x,y, 'c--')
```

```python
plt.plot(x,y)
plt.xlabel("x title")
plt.ylabel("y title")
plt.title("This is ploting")
plt.xticks(range(0,18,3))
plt.show()
```

```python
plt.subplot(1,3,1)
plt.plot(x,y)

plt.subplot(1,3,2)
plt.plot(x,y,'r--')

plt.subplot(1,3,3)
plt.plot(x,y,'c--')

plt.tight_layout()
```

```python
fig1 = plt.figure(figsize=(8,5))
axis = fig1.add_axes([.1,.1,1,1])
plt.plot(x,y,'r--')
axis.set_xlabel("x Label")
axis.set_ylabel("y label")
axis.set_title("object oriented")
fig1.savefig('savephoto.pdf', dpi = 200)
```

```python
fig,axis = plt.subplots(2,2, sharex = True,sharey = True,figsize = (15,5))
axis[0][0].plot(x,y)
axis[0][1].plot(x,y)
axis[1][0].plot(x,y)
axis[1][1].plot(x,y)
```
